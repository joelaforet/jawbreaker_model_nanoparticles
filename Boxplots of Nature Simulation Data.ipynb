```python
from scipy import stats
```

```python
x =  [x for x in df2.columns[2:-1]]

for i in x:
    #print(i)
    k2, p = stats.normaltest(df2[i])

    alpha = 0.05
    print("p = {:g}".format(p))

    if p < alpha: # null hypothesis: x comes from a normal distribution
        print("The null hypothesis can be rejected for {}".format(i))
    else:
         print("The null hypothesis cannot be rejected for {}".format(i))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
feature_names = df2.columns[2:-1]
df_long = pd.melt(df2, 'CLASS', var_name='Feature', value_name='Value') 

all_t = list()
all_p = list()
for case in range(len(feature_names)):
    sub_df = df_long[df_long.Feature == feature_names[case]]
    g1 = sub_df[sub_df['CLASS'] == 0]['Value'].values
    g2 = sub_df[sub_df['CLASS'] == 1]['Value'].values
    t, p = stats.ttest_ind(g1, g2)
    all_t.append(t)
    all_p.append(p)
    
    t, p = stats.ttest_ind(g1, g2)
    
#print(all_t)
#print(all_p)
print(feature_names)
print(np.count_nonzero(np.array(feature_names)[np.array(all_p) < 0.05]))

```

```python
# renaming so that class 0 will appear as setosa and class 1 as versicolor
df_long.loc[df_long.CLASS==0, 'CLASS'] = 'Negative'
df_long.loc[df_long.CLASS==1, 'CLASS'] = 'Positive'
# Boxplots
fig, axes = plt.subplots(2,2, figsize=(20,10), dpi=300)
axes = axes.flatten()
for idx, feature in enumerate(feature_names):
    ax = sns.boxplot(x='Feature', hue='CLASS', y='Value', data = df_long[df_long.Feature == feature], linewidth=2, showmeans=True, meanprops={'marker':'*','markerfacecolor':'white', 'markeredgecolor':'black'}, ax=axes[idx])
    #* tick params
    axes[idx].set_xticklabels([str(feature)], rotation=0)
    axes[idx].set(xlabel=None)
    axes[idx].set(ylabel=None)
    axes[idx].grid(alpha=0.5)
    axes[idx].legend(loc='lower right', prop={'size': 11})
 
    #*set edge color = black
    for b in range(len(ax.artists)):
        ax.artists[b].set_edgecolor('black')
        ax.artists[b].set_alpha(0.8)
   
    #* statistical tests
    x1, x2 = -0.20, 0.20
    y, h, col = df_long[df_long.Feature == feature]['Value'].max()+1, 2, 'k'
    axes[idx].plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
    axes[idx].text((x1+x2)*.5, y+h, 'statistically significant', ha='center', va='bottom', color=col)
fig.suptitle("Significant feature differences between positive and negative nanoparticle samples", size=14, y=0.93)
plt.show()
```

```python
def convert(x):
    if x == 0:
        return 'Negative'
    else:
        return 'Positive'
```
