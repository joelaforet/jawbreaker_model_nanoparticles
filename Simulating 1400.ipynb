# Clean names of Drugs/Excips

```python
def cleanNames(col):
    col = col.capitalize()    
    split = col.split('_')
    for a in range(len(split)):
        split[a] = split[a].capitalize()
    return ''.join(split)
```

# Generate Job files for making mols

```python
def writeMolJob(molFile, jobNum):
    
    
    jobName = "MakeMol_"+ molFile.split('.')[0]
    job_file = os.path.join(os.getcwd(), '1MolJob%s.sh' % str(jobNum))
    
    with open(job_file, 'w') as fh:
        fh.write("#!/bin/bash\n")
        fh.write('#SBATCH --mail-type=begin\n')
        fh.write('#SBATCH --mail-type=end\n')
        fh.write('#SBATCH --mail-user=jrl78@duke.edu\n')
        fh.write('#SBATCH -e slurm.err\n')
        fh.write("#SBATCH --job-name=%s.job\n" % jobName)
        fh.write("#SBATCH --mem=20G\n")
        fh.write("#SBATCH -p scavenger-gpu --gres=gpu:1\n")
        fh.write("#SBATCH --exclusive\n")
        
        fh.write('./make_mol.sh ManySims/molFiles/%s\n' % molFile)
       
    
    
    
```
