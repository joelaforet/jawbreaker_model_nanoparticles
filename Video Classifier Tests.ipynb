```python
def frames_extraction(simFolder):
    #List to store the end normalized resized frames
    frames_list = []
    
    # Go into the input folder
    # Get all of the images as np Arrays
    # Exit Folder and return list
    curDir = os.getcwd()
    os.chdir(simFolder)
    
    frames = glob.glob("*.png")
    for x in frames:
        img = cv2.imread(x, cv2.IMREAD_UNCHANGED)
        #resized_frame = cv2.resize(img, (112,112))
        
        #Normalize resized frame to get pixel values between 0 and 1
        normalized_frame = img/255
        
        #Appends normalized frame to return list
        frames_list.append(normalized_frame)
    
    return frames_list
    
```

```python
def create_dataset():
    
    #Declare Empty Lists to store the features and label values
    temp_features = []
    features = []
    labels = []
    
    for class_index, class_name in enumerate(classes_list):
        print(f'Extracting Data of Class: {class_name}')
        curDir = os.path.join(dataset_directory, class_name)
        
        
        #Get the list of sims in the directory
        folder_list = os.listdir(curDir)
        os.chdir(curDir)
        #Iterate over every folder in the class Folder
        
        for folder_name in folder_list:
            
            sim_file_path = os.path.join(dataset_directory, class_name, folder_name)
            
            # Call the frame_extraction on each folder of sims to get the np array
            name = folder_name.split('_')[2] + '_' + folder_name.split('_')[3]+ '_12_4_'
            
            frames = frames_extraction(name)
            os.chdir(curDir)
            # Append these np representations to a temporary list
        print(temp_features)
        print(max_images_per_class)
        features.extend(random.sample(temp_features, max_images_per_class))
        
        #Add fixed number of labels to the labels list
        labels.extend([class_index] * max_images_per_class)
            
        # Empty temp_features list for reuse
        temp_features.clear()
            
        # Convert features and labels lists to numpy arrays
        features = np.asarray(features)
        labels = np.array(labels)
            
        return features, labels
```
